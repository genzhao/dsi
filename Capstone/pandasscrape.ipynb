```python
frames = [zappos,six]
```

```python
zapposmerged = pd.concat(frames, axis=0, ignore_index=True)
```

```python
zapposmerged.brand_name.value_counts()
```

```python
zappos_clean = zapposmerged.groupby('brand_name').filter(lambda x: len(x) > 100).reset_index(drop=True)
```

```python
test_list = []
for i in zappos_clean.lateralfrontleft:
    test_list.append('http://' + i)
```

```python
zappos_clean['lateralfrontleft'] = test_list
```

```python
zappos_clean.to_csv('zappos_clean.csv', index=False)
```

```python
temp_list = []
for i in nike['shoe_name']:
    if 'Nike' in i:
        temp_list.append('Nike')
    elif 'Jordan' in i :
        temp_list.append('Jordan')
    elif 'Converse' in i:
        temp_list.append('Converse')
    elif 'LeBron' in i:
        temp_list.append('Nike')
    elif 'Kobe' in i:
        temp_list.append('Nike')
    elif 'Kyrie' in i:
        temp_list.append('Nike')
    elif 'Jack Purcell' in i:
        temp_list.append('Converse')
    elif 'Chuck Taylor' in i:
        temp_list.append('Converse')
    else:
        temp_list.append('Nike')
```

```python
nike['brand_name'] = temp_list
```
